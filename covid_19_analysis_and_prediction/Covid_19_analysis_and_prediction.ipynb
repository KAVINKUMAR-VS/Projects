**Joining the data we need in the two dataframes into a single dataframe**

```python
code=data_transformed["CODE"].unique().tolist()
country=data_transformed["COUNTRY"].unique().tolist()
hdi=[]
tc=[]
td=[]
sti=[]
population=data_transformed["POP"].unique().tolist()
gdp=[]

for i in country:
    hdi.append((data_transformed.loc[data_transformed["COUNTRY"]==i, "HDI"]).sum()/294)
    tc.append((data_raw.loc[data_raw["location"]==i, "total_cases"]).sum())
    td.append((data_raw.loc[data_raw["location"]==i, "total_deaths"]).sum())
    sti.append((data_transformed.loc[data_transformed["COUNTRY"]==i, "STI"]).sum()/294)
    population.append((data_raw.loc[data_raw["location"]==i, "population"]).sum()/294)

aggregated_data = pd.DataFrame(list(zip(code, country, hdi, tc, td, sti, population)),
                               columns = ["Country Code", "Country", "HDI",
                                          "Total Cases", "Total Deaths",
                                          "Stringency Index", "Population"])
display(aggregated_data.head())
```

**Sorting data_transformed dataframe in descending order by Total Cases**

```python
data_transformed=aggregated_data.sort_values(by=["Total Cases"], ascending=False)
display(data_transformed.head())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from datetime import timedelta

raw_data = pd.read_csv('raw_data.csv')

country_name = input("Enter the country name: ")

country_data = raw_data[raw_data['location'] == country_name]

country_data = country_data.sort_values(by='date')

country_data['date'] = pd.to_datetime(country_data['date'])

country_data['days_since_first_case'] = (country_data['date'] - country_data['date'].min()).dt.days

X = country_data[['days_since_first_case']]
y = country_data['total_cases'].fillna(country_data['total_cases'].mean())

model = LinearRegression()
model.fit(X, y)

print(model.coef_,model.intercept_)

last_date = country_data['date'].max()
next_25_dates = [last_date + timedelta(days=i) for i in range(1, 26)]
days_since_first_case = [(date - country_data['date'].min()).days for date in next_25_dates]
predicted_cases = model.predict(np.array(days_since_first_case).reshape(-1, 1))

print("Current date as per dataset: ", last_date)
print("\n")

predicted_data = pd.DataFrame({'date': next_25_dates, 'predicted_cases': predicted_cases})

print(predicted_data)

#Afghanistan
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=predicted_data, x='date', y='predicted_cases', marker='o', markersize=5, color='red', linewidth=2, label='Predicted COVID-19 Cases')
```
